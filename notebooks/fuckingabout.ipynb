```python
# read training image metadata csv

import pandas as pd
import numpy as np

df = pd.read_csv('split_classes.csv')

df['red_ave'] = np.NaN
df['green_ave'] = np.NaN
df['blue_ave'] = np.NaN
df['red_var'] = np.NaN
df['green_var'] = np.NaN
df['blue_var'] = np.NaN

# update date
```

```python
# select relevant part of image using mask, and associate the resulting array with the appropriate record from the csv

import os
import pickle
from PIL import Image, ImageDraw
import numpy as np
from matplotlib.pyplot import imshow
import time

masks = pickle.load(open('masks.p', 'rb'))

image_dir = 'images'
c = 0
l = len(df)
for image_file in df['filename']:
    print(f'{c}/{l}'); c += 1
    if (image_file[-4:] == '.jpg'):
        
        # read image as array
        pil_im = Image.open(f'{image_dir}/{image_file}', 'r')
        arr = np.asarray(pil_im)
        #print(arr.shape)
        
        # mask image
        cam = image_file.split('_')[0]
        mask = masks[cam]
        
        reds = []
        blues = []
        greens = []
        # imshow(np.dstack((arr[:,:,0], np.full(arr.shape[:2], 0), np.full(arr.shape[:2], 0))))
        
        for i in range(mask.shape[0]):
            for j in range(mask.shape[1]):
                if mask[i][j] != 0:
                    reds.append(arr[i][j][0])
                    greens.append(arr[i][j][1])
                    blues.append(arr[i][j][2])
                 
        reds = np.array(reds)
        blues = np.array(blues)
        greens = np.array(greens)
        
        df.loc[df.filename == image_file, 'red_ave'] = reds.mean()
        df.loc[df.filename == image_file, 'green_ave'] = greens.mean()
        df.loc[df.filename == image_file, 'blue_ave'] = blues.mean()
        
        df.loc[df.filename == image_file, 'red_var'] = reds.var()
        df.loc[df.filename == image_file, 'green_var'] = greens.var()
        df.loc[df.filename == image_file, 'blue_var'] = blues.var()
           
        
    
```

```python
# write df to new csv
display(df)
df.to_csv('stats.csv')
```
