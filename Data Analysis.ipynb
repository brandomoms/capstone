```python
import pandas as pd
import numpy as np

df = pd.read_csv('stats.csv')
```

```python
# plot number of images per class (how representative is the data set?)
%matplotlib notebook
import matplotlib.pyplot as plt

classes = ['TRACK', 'SNOW_SLUSH', 'ICE', 'CLEAR', 'UNK']
class_counts = [
    len(df[df['TRACK'] == True]), 
    len(df[df['SNOW_SLUSH'] == True]), 
    len(df[df['ICE'] == True]), 
    len(df[df['CLEAR'] == True]), 
    len(df[df['UNK'] == True])]

plt.bar(classes, class_counts)
#print()
```
