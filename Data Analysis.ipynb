```python
import pandas as pd
import numpy as np

df = pd.read_csv('stats.csv')
df = df[df['UNK'] == False]
display(df)
```

```python
# plot number of images per class (how representative is the data set?)
%matplotlib notebook
import matplotlib.pyplot as plt

classes = ['TRACK', 'SNOW_SLUSH', 'ICE', 'CLEAR', 'UNK']
class_counts = [
    len(df[df['TRACK'] == True]), 
    len(df[df['SNOW_SLUSH'] == True]), 
    len(df[df['ICE'] == True]), 
    len(df[df['CLEAR'] == True]), 
    len(df[df['UNK'] == True])]

plt.bar(classes, class_counts)
#print()
```

```python
import matplotlib.pyplot as plt

classes = ['TRACK', 'SNOW_SLUSH', 'ICE', 'CLEAR']

red_mean_df = df[['TRACK', 'SNOW_SLUSH', 'ICE', 'CLEAR', 'red_ave']]
green_mean_df = df[['TRACK', 'SNOW_SLUSH', 'ICE', 'CLEAR', 'green_ave']]
blue_mean_df = df[['TRACK', 'SNOW_SLUSH', 'ICE', 'CLEAR', 'blue_ave']]

f = plt.figure()    
f, axes = plt.subplots(nrows=1, ncols=3, sharex=False, sharey=True)
f.subplots_adjust(wspace=0.2)
plt.xticks(range(len(classes)), classes, rotation=75)

axes[0].scatter([1]*len(red_mean_df[red_mean_df['TRACK'] == True]), red_mean_df[red_mean_df['TRACK'] == True]['red_ave'])
axes[0].scatter([2]*len(red_mean_df[red_mean_df['SNOW_SLUSH'] == True]), red_mean_df[red_mean_df['SNOW_SLUSH'] == True]['red_ave'])
axes[0].scatter([3]*len(red_mean_df[red_mean_df['ICE'] == True]), red_mean_df[red_mean_df['ICE'] == True]['red_ave'])
axes[0].scatter([4]*len(red_mean_df[red_mean_df['CLEAR'] == True]), red_mean_df[red_mean_df['CLEAR'] == True]['red_ave'])
axes[0].set_title('Red Channel')
axes[0].set_xticks(range(1, len(classes)+1))
axes[0].set_xticklabels(classes, rotation=75)
axes[0].set_ylabel('per image mean pixel value')

axes[1].scatter([1]*len(green_mean_df[green_mean_df['TRACK'] == True]), green_mean_df[green_mean_df['TRACK'] == True]['green_ave'])
axes[1].scatter([2]*len(green_mean_df[green_mean_df['SNOW_SLUSH'] == True]), green_mean_df[green_mean_df['SNOW_SLUSH'] == True]['green_ave'])
axes[1].scatter([3]*len(green_mean_df[green_mean_df['ICE'] == True]), green_mean_df[green_mean_df['ICE'] == True]['green_ave'])
axes[1].scatter([4]*len(green_mean_df[green_mean_df['CLEAR'] == True]), green_mean_df[green_mean_df['CLEAR'] == True]['green_ave'])
axes[1].set_title('Green Channel')
axes[1].set_xticks(range(1, len(classes)+1))
axes[1].set_xticklabels(classes, rotation=75)

axes[2].scatter([1]*len(blue_mean_df[blue_mean_df['TRACK'] == True]), blue_mean_df[blue_mean_df['TRACK'] == True]['blue_ave'])
axes[2].scatter([2]*len(blue_mean_df[blue_mean_df['SNOW_SLUSH'] == True]), blue_mean_df[blue_mean_df['SNOW_SLUSH'] == True]['blue_ave'])
axes[2].scatter([3]*len(blue_mean_df[blue_mean_df['ICE'] == True]), blue_mean_df[blue_mean_df['ICE'] == True]['blue_ave'])
axes[2].scatter([4]*len(blue_mean_df[blue_mean_df['CLEAR'] == True]), blue_mean_df[blue_mean_df['CLEAR'] == True]['blue_ave'])
axes[2].set_title('Blue Channel')
axes[2].set_xticks(range(1, len(classes)+1))
axes[2].set_xticklabels(classes, rotation=75)

plt.show()
```
